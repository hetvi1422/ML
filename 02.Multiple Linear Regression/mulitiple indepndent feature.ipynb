```python
X = df[['Temperature_C', 'Humidity_%', 'Weekend', 'Holiday', 'Promotion', 'Footfall']]
y = df['IceCream_Sales']
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python

plt.scatter(y, y_pred)
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Multiple Linear Regression: Actual vs Predicted")
plt.show()
```

```python
plt.scatter(y, y_pred, label='Data Points')

# Add regression line
m, b = np.polyfit(y, y_pred, 1)
plt.plot(y, m*y + b, color='blue', linewidth=1, label='Best Fit Line')  # DARK BLUE LINE

plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual vs Predicted with Line")
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sort values for smooth lines
sorted_index = np.argsort(y.values)

y_sorted = y.values[sorted_index]
y_pred_sorted = y_pred[sorted_index]

# Plot lines
plt.plot(y_sorted, label='Actual Sales', color='blue', linewidth=4)   # DARK BLUE
plt.plot(y_pred_sorted, label='Predicted Sales', color='orange', linewidth=2)

plt.xlabel("Data Points")
plt.ylabel("IceCream Sales")
plt.title("Actual vs Predicted (Line Chart)")
plt.legend()

plt.show()
```
