```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Lets create synthetic data points
from sklearn.datasets import make_classification
```

```python
# no. of cluster per class is 2
X,y=make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0, random_state=42)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
# for this type of data points use rbf kernal 
rbf = SVC(kernel='rbf', gamma='scale')
```

```python
# train the model using rbg kernel 
rbf.fit(X_train, y_train)
```

```python
##Prediction
y_pred1 = rbf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1)) 
# we got 89 percent accuracy using rbf kernel  
```

```python
rbf.intercept_
```

```python
# now train the model using polynomial kernel 
polynomial = SVC(kernel='poly', degree=3)
polynomial.fit(X_train, y_train)
```

```python
# Prediction
y_pred2 = polynomial.predict(X_test)
```

```python
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2)) 
# we got 88% accuracy using polynomial kernel 
```

```python
polynomial.intercept_
```

```python
# now train the model using polynomial kernel 
sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
```

```python
# Prediction
y_pred3=sigmoid.predict(X_test)
```

```python
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3)) 
# we got 74% accuracy using sigmoid kernel 
```

```python
sigmoid.intercept_
```

## Hyperparameter tunnig with SVC using GridsearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid=GridSearchCV(SVC(), param_grid=param_grid, refit=True, cv=5, verbose=3)
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
```

```python
rbf = SVC(kernel='rbf',gamma='scale')
rbf.fit(X_train, y_train)
```

```python
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
# we got 92% accuracy by doing hyperparameter tunnig 
```
